```python
for i in range(1, len(sales)):

    growth = ((sales[i] - sales[i-1]) / sales[i-1]) * 100

    print(f"{quarters[i-1]} → {quarters[i]} = {growth:.2f}%")
```

```python
plt.figure(figsize=(8,5))

plt.plot(quarters, sales, marker='o')

plt.title("Zomato Revenue Growth")

plt.xlabel("Quarter")

plt.ylabel("Revenue (Rs. Crores)")

plt.grid(True)

plt.show()
```
